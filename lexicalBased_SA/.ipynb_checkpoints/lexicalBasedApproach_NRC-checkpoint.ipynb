# Lexic for news

```python
import csv
import pandas as pd
from nltk.stem import WordNetLemmatizer

path = "NRC-Sentiment-Emotion-Lexicons/NRC-Sentiment-Emotion-Lexicons/NRC-Affect-Intensity-Lexicon/NRC-AffectIntensity-Lexicon.txt"
```

```python
with open(path,'r') as tsvin, open('NRC-AffectIntensity-Lexicon.csv', 'w') as csvout:
    tsvin = csv.reader(tsvin, delimiter='\t')    
    csvout = csv.writer(csvout)
    for row in tsvin:
        csvout.writerow(row)
```

```python
lexicon = pd.read_csv("NRC-AffectIntensity-Lexicon.csv")
lexicon
```

```python
en_news = pd.read_pickle("../data/final/NotIot_clean_EN_news.p")
en_news
```

```python
from nltk.tokenize import word_tokenize

lexicon_feature = pd.DataFrame() 
lema = WordNetLemmatizer()

for s in en_news["clean_text"]:
    joy = 0 
    sadness = 0 
    fear = 0 
    anger = 0 
    print(s)
    for w in word_tokenize(s): 
        count_word = (len(word_tokenize(s)))
        if  w in set(lexicon['term']):
            print(w)
            affect = set(lexicon[(lexicon['term']==w)]['AffectDimension'])
            for aff in affect: 
                if aff=='anger':
                    anger = anger + float(lexicon[(lexicon['term']==w) & (lexicon['AffectDimension']==aff) ]['score'])
                    print('this is anger ')
                   # print(anger)
                if aff=='fear':
                    fear = fear + float(lexicon[(lexicon['term']==w) & (lexicon['AffectDimension']==aff) ]['score'])
                    print('this is fear ')
                    #print(fear)
                if aff=='sadness':
                    sadness = sadness + float(lexicon[((lexicon['term']==w)) & (lexicon['AffectDimension']==aff) ]['score'])
                    print('this is sadness ')
                  #  print(sadness)
                if aff=='joy':
                    joy = joy + float(lexicon[(lexicon['term']==w) & (lexicon['AffectDimension']==aff) ]['score'])
                    print('this is joy ')
                   # print(joy)
        
        elif lema.lemmatize(w) in set(lexicon['term']): 
            print(lema.lemmatize(w))
            affect = set(lexicon[(lexicon['term']==lema.lemmatize(w))]['AffectDimension'])
            for aff in affect: 
                if aff=='anger':
                    anger = anger + float(lexicon[((lexicon['term']==lema.lemmatize(w))) & (lexicon['AffectDimension']==aff) ]['score'])
                    print('this is anger ')
                   # print(anger)
                if aff=='fear':
                    fear = fear + float(lexicon[((lexicon['term']==lema.lemmatize(w))) & (lexicon['AffectDimension']==aff) ]['score'])
                    print('this is fear ')
                    #print(fear)
                if aff=='sadness':
                    sadness = sadness + float(lexicon[((lexicon['term']==lema.lemmatize(w))) & (lexicon['AffectDimension']==aff) ]['score'])
                    print('this is sadness ')
                  #  print(sadness)
                if aff=='joy':
                    joy = joy + float(lexicon[((lexicon['term']==lema.lemmatize(w))) & (lexicon['AffectDimension']==aff) ]['score'])
                    print('this is joy ')
                   # print(joy)
        
                
    joy = joy / count_word 
    anger = anger / count_word 
    sadness = sadness / count_word 
    fear = fear / count_word 
    lexicon_feature = lexicon_feature.append({'lex_joy': joy, 'lex_fear': fear,'lex_anger':anger,'lex_sadness':sadness}, ignore_index=True)
```

```python
en_news.index = lexicon_feature.index
en_news['lex_anger'] = lexicon_feature['lex_anger']
en_news['lex_sadness'] = lexicon_feature['lex_sadness']
en_news['lex_joy'] = lexicon_feature['lex_joy']
en_news['lex_fear'] = lexicon_feature['lex_fear']
#print( lexicon_feature['lex_fear'])
#len(lexicon_feature)
```

```python
en_news.to_pickle('../data/final/NotIot_clean_EN_news_lexic.p') # save dframe to pickle
```
