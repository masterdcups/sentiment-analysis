# News Lexic

```python
from xml.dom import minidom
import pandas as pd
from french_lefff_lemmatizer.french_lefff_lemmatizer import FrenchLefffLemmatizer
lemmatizer = FrenchLefffLemmatizer()
```

```python
lexicon = pd.DataFrame() 

mydoc = minidom.parse('Lexique_EMOTAIX_structure.xml')
```

```python
items = mydoc.getElementsByTagName('Entry')
```

```python
for i in items: 
    lema = i.getElementsByTagName('lemma')[0].firstChild.nodeValue
    print(lema)
    sens = i.getElementsByTagName('sense')[0].attributes['category'].value
    print(sens)
    lexicon = lexicon.append({'term': lema, 'sense': sens}, ignore_index=True)
```

```python
lexicon
```

```python
lexicon.groupby('sense').count()
```

```python
lexicon['sense'][lexicon['sense']=='Bien_Etre'] = 'joy'
lexicon['sense'][lexicon['sense']=='Bienveillance'] = 'joy'
lexicon['sense'][lexicon['sense']=='Sang_froid'] = 'joy'
lexicon['sense'][lexicon['sense']=='anxiete'] = 'fear'
lexicon['sense'][lexicon['sense']=='emotion_non_specifiee'] = 'neutre'
lexicon['sense'][lexicon['sense']=='mal_être'] = 'sadness'
lexicon['sense'][lexicon['sense']=='malveillance'] = 'anger'
lexicon['sense'][lexicon['sense']=='surprise'] = 'surprise'
```

```python
lexicon_feel = pd.read_csv('FEEL.csv', sep=';', header =0, encoding='utf-8')
lexicon_feel
```

```python
fr_news = pd.read_pickle("../data/final/NotIot_clean_FR_news.p")
fr_news
```

```python
from nltk.tokenize import word_tokenize

lexicon_feature = pd.DataFrame() 

for s in fr_news["clean_text"]:
    joy_EMO = 0 
    sadness_EMO = 0 
    fear_EMO = 0 
    anger_EMO = 0 
    surprise_EMO = 0 
    neutre_EMO = 0
    
    joy_feel = 0 
    sadness_feel = 0 
    fear_feel = 0 
    anger_feel = 0 
    surprise_feel = 0 
    disgust_feel = 0 
    
    joy = 0 
    sadness = 0 
    fear = 0 
    anger = 0 
    surprise = 0 
    disgust = 0 
    neutre = 0 
    print(s)
    #lexicon EMOTAIX
    for w in word_tokenize(s): 
        count_word = (len(word_tokenize(s)))
        if  w in set(lexicon['term']):
            print(w)
            affect = set(lexicon[(lexicon['term']==w)]['sense'])
            for aff in affect: 
                if aff=='anger':
                    anger_EMO = anger_EMO + 1
                    print('this is anger ')
                   # print(anger)
                if aff=='fear':
                    fear_EMO = fear_EMO + 1
                    print('this is fear ')
                    #print(fear)
                if aff=='sadness':
                    sadness_EMO = sadness_EMO + 1
                    print('this is sadness ')
                  #  print(sadness)
                if aff=='joy':
                    joy_EMO = joy_EMO + 1
                    print('this is joy ')
                   # print(joy)
                if aff=='neutre':
                    neutre_EMO = neutre_EMO + 1
                    print('this is neutre ')
                   # print(joy)
                if aff=='surprise':
                    surprise_EMO = surprise_EMO + 1
                    print('this is surprise ')
        
        elif lemmatizer.lemmatize(w) in set(lexicon['term']): 
            print(lemmatizer.lemmatize(w))
            affect = set(lexicon[(lexicon['term']==lemmatizer.lemmatize(w))]['sense'])
            for aff in affect: 
                if aff=='anger':
                    anger_EMO = anger_EMO + 1
                    print('this is anger ')
                   # print(anger)
                if aff=='fear':
                    fear_EMO = fear_EMO + 1
                    print('this is fear ')
                    #print(fear)
                if aff=='sadness':
                    sadness_EMO = sadness_EMO + 1
                    print('this is sadness ')
                  #  print(sadness)
                if aff=='joy':
                    joy_EMO = joy_EMO + 1
                    print('this is joy ')
                   # print(joy)
                if aff=='neutre':
                    neutre_EMO = neutre_EMO + 1
                    print('this is neutre ')
                   # print(joy)
                if aff=='surprise':
                    surprise_EMO = surprise_EMO + 1
                    print('this is surprise ')
                    
    for w in word_tokenize(s): 
    #Lexicon FEEL
        word = lemmatizer.lemmatize(w)
        if word in set(lexicon_feel["word"]):
            #score = score + lexicon[word]
            i = list(lexicon_feel["word"]).index(word)
            print(word)
            #print(lexicon_feel["joy"])
            joy_feel = joy_feel + lexicon_feel["joy"][i]
            #print(lexicon_feel["fear"])
            fear_feel = fear_feel + lexicon_feel["fear"][i]
            #print(lexicon_feel["sadness"])
            sadness_feel = sadness_feel+ lexicon_feel["sadness"][i]
            #print(lexicon_feel["anger"])
            anger_feel = anger_feel + lexicon_feel["anger"][i] 
            #print(lexicon_feel["surprise"])
            surprise_feel = surprise_feel + lexicon_feel["surprise"][i]
            #print(lexicon_feel["disgust"])
            disgust_feel = disgust_feel + lexicon_feel["disgust"][i] 
    
    surprise = max(surprise_feel, surprise_EMO) / count_word            
    joy = max(joy_feel,joy_EMO) / count_word 
    anger = max(anger_feel , anger_EMO) / count_word 
    sadness = max(sadness_feel ,sadness_EMO) / count_word 
    fear = max(fear_feel , fear_EMO) / count_word 
    neutre = neutre_EMO / count_word
    disgust = disgust_feel / count_word
    lexicon_feature = lexicon_feature.append({'lex_neutre':neutre,'lex_disgust':disgust,'lex_surprise':surprise,'lex_joy': joy, 'lex_fear': fear,'lex_anger':anger,'lex_sadness':sadness}, ignore_index=True)
```

```python
lexicon_feature
```

```python
fr_news.index = lexicon_feature.index
fr_news['lex_anger'] = lexicon_feature['lex_anger']
fr_news['lex_sadness'] = lexicon_feature['lex_sadness']
fr_news['lex_joy'] = lexicon_feature['lex_joy']
fr_news['lex_fear'] = lexicon_feature['lex_fear']
fr_news['lex_surprise'] = lexicon_feature['lex_surprise']
fr_news['lex_disgust'] = lexicon_feature['lex_disgust']
fr_news['lex_neutre'] = lexicon_feature['lex_neutre']
```
