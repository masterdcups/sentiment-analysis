```python
#generation des ngram 
for review in data['text']:
    s = review.lower()
    #s = re.sub(r'[^a-zA-Z0-9\s]', ' ', s)
    tokens = [token for token in s.split(" ") if token != ""]
    output = list(ngrams(tokens, 2))
    print(output)
```

```python
for review in data['text']:
    score = 0
    for word in review.split():
        if word in lexicon:
            score = score + lexicon[word]
```
